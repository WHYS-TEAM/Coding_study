```python
# https://school.programmers.co.kr/learn/courses/30/lessons/133500
def check_num_off(status, graph, node):
    cnt = 0
    if status[node] == 0:
        cnt += 1
    for end_node in graph[node]:
        if status[end_node] == 0:
            cnt += 1
    return cnt

def solution(n, lighthouse):
    if n == 2:
        return 1
    answer = 0
    status = [-1] + [0]*(n)
    connect_num = [-1] + [0]*(n)
    graph = [[] for _ in range(n+1)]
    
    # graph 생성
    for i in range(len(lighthouse)):
        start_node = lighthouse[i][0]
        end_node = lighthouse[i][1]
        graph[start_node].append(end_node)
        graph[end_node].append(start_node)
    # connect_num 꺼진 등대 연결 개수
    for i in range(1, n+1):
        if len(graph[i]) == 1:
            connect_num[i] = 0
        else:
            connect_num[i] = len(graph[i])
        
    while True:
        max_num_off = 0
        target_node = -1
        for node in range(1,n+1):
            #if complete[node] == 0:
            if connect_num[node] > 0:
                num_off = connect_num[node]
                # if num_off == 0:
                #     complete[node] = 1
                if num_off > max_num_off:
                    target_node = node
                    max_num_off = num_off
        # 끝
        if max_num_off == 0:
            break
        # 등대 켜기
        status[target_node] = 1
        # complete[target_node] = 1
        connect_num[target_node] = 0
        for end_node in graph[target_node]:
            # status[end_node] = 1
            connect_num[end_node] = connect_num[end_node] - 1
            
        answer += 1

    return answer
```
